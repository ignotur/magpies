```python
import sys
sys.path.insert(1, '../src/magpies/')
from magpies import *
from atmos import *
import numpy as np
from math import *
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
## Radius and mass of neutron star
## (a) panel 
Rns = 10  ## km
Mns = 1.8 ## M_solar
Tb = 1e9  ## K 
Bp = 1e13 ## G

g14c = g14 (Rns, Mns)
```

```python
atm_iron_2003 = NS_atmosphere ('Potekhin_2003_iron', g14c, Tb, Bp)
Tm = Tmap (usage='NS_atm', ns_atm=atm_iron_2003, Ntheta = 320, Nphi=320)

Tm.plot_Ts(filename='surface_temperature_9_13.png')
```

```python
phases = np.linspace (0, 4*pi, 400) ## Phases where we compute the lightcurve

## chi - angle between rotational axis and orientation of the magnetic dipole
## inc - inclination of the observer with respect to rotational axis

## Let us look at different locations of the observer
## Since the thermal map is axisymmetric we have to introduce some angle between rotational axis and orientation of the dipolar axis

intens_0 = lightcurve (Tm, Rns, Mns, phases, pi/2, 0) ## last two arguments are chi and inclination
intens_1 = lightcurve (Tm, Rns, Mns, phases, pi/2, pi/2) ## last two arguments are chi and inclination
```

```python
plt.plot (phases, intens_0, 'b--', label=r'$\chi = \pi/2$ and $i = 0$')
plt.plot (phases, intens_1, 'r:', label=r'$\chi = \pi/2$ and $i = \pi/2$')
#plt.ylim ([3.62e33, 3.76e33])
plt.legend(loc=1)
```

```python
intens_2 = lightcurve (Tm, Rns, Mns, phases, 0, 0) ## last two arguments are chi and inclination
intens_3 = lightcurve (Tm, Rns, Mns, phases, 0, pi/2) ## last two arguments are chi and inclination
```

```python
plt.plot (phases, intens_2, 'b--', label=r'$\chi=0$ and $i = 0$')
plt.plot (phases, intens_3, 'r:', label=r'$\chi=0$ and $i = pi/2$')
#plt.ylim ([3.62e33, 3.76e33])
plt.legend(loc=1)
```

```python
theta1, phi1 = np.meshgrid (Tm.theta, Tm.phi)

val = precompute_Dcos_alpha (Rns, Mns, 0, pi/2, 0, phi1, theta1)
bc = plt.imshow(val.T)
plt.colorbar(bc)
print (np.max(val), np.min(val))
```
