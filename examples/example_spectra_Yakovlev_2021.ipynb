```python
import sys
sys.path.insert(1, '../src/magpies/')
from magpies import *
from atmos import *
import numpy as np
from math import *
from scipy.optimize import minimize
import matplotlib as mpl
import matplotlib.pyplot as plt
```

Producing figures from article by Yakovlev (2021) 2021MNRAS.506.4593Y
Some figures look differently because of the typos in that article

```python
## Radius and mass of neutron star
Rns = 12  ## km
Mns = 1.4 ## M_solar
Tb = pow(10, 7.1730)  ## K
Bp = 1e14 ## G

g14 = g14 (Rns, Mns) ## computing the free fall acceleration
 
print ('g14 = ', g14)
```

```python
## Initialising the atmosphere
atm_iron_2003 = NS_atmosphere ('Potekhin_2003_iron', g14, Tb, Bp)
atm_iron_2003.describe ()
```

```python
theta = np.linspace (0, pi, 100)  ## theta coordinates
phi   = np.linspace (0, 2*pi, 99) ## phi coordinates

theta1, phi1 = np.meshgrid (theta, phi)

Ts = atm_iron_2003.Ts (theta1) ## Surface temperatures
```

```python
Teff = compute_Teff (theta, phi, Rns, Ts)
print ('log_10 Teff = ', round(log10(Teff), 1))
```

```python
## Here we plot the surface thermal map to visually inspect the result

frame = plt.subplot(111, projection='aitoff')
bc = plt.contourf (phi-pi, -(theta-pi/2), Ts.T, 40)
#plt.colorbar(bc)
frame.axes.xaxis.set_ticklabels([])
plt.tight_layout()
```

```python
plt.plot (theta, np.log10(Ts[0]))
plt.xlabel(r'$\theta$')
plt.ylabel(r'$\log_{10} T_s$')
```

```python
## Here we produce and store the soft X-ray spectra for two orientations: pole on and equatorial
eph, sp_pole, map_of_visible_pole = get_redshifted_spectra_pole_3D (theta, phi, Ts, Rns, Mns)
eph, sp_eq,   map_of_visible_eq   = get_redshifted_spectra_equator_3D (theta, phi, Ts, Rns, Mns)
## Single black body spectra with Teff
eph, bb1 = single_BB (Teff, Rns, Mns)
```

```python
frame = plt.subplot(111, projection='aitoff')
bc = plt.contourf (phi-pi, -(theta-pi/2), map_of_visible_pole.T, 40)
#plt.colorbar(bc)
frame.axes.xaxis.set_ticklabels([])
plt.tight_layout()

## Dark color - effective temperature 0 shows area which does not contribute to the spectra because it is invisible for the observer
```

```python
frame = plt.subplot(111, projection='aitoff')
bc = plt.contourf (phi-pi, -(theta-pi/2), map_of_visible_eq.T, 40)
#plt.colorbar(bc)
frame.axes.xaxis.set_ticklabels([])
plt.tight_layout()
```

```python
## A factor of 4 difference in absolute value caused by correcting the typo in eq. (4) of Yakovlev
## Roughly corresponds to Figure 2 in Yakovlev (2021) article

plt.plot (np.log10(eph), np.log10(sp_pole), 'r-',  label=r'Pole red-shifted')
plt.plot (np.log10(eph), np.log10(sp_eq), 'k--',  label=r'Equator red-shifted')
plt.plot (np.log10(eph), np.log10(bb1), 'b:',  label=r'$\log_{10} T_\mathrm{eff}=$'+str(round(log10(Teff), 1)))



plt.ylim([15,23])
plt.xlim([-1.2, 0.7])
plt.xlabel(r'$\log\; E$ (keV)')
plt.ylabel(r'$\log\;  H_E^\infty$ (erg s$^{-1}$ cm$^{-2}$ keV$^{-1}$)')
plt.tight_layout()
plt.legend()
```

```python
## Here we suggest an initial guess for a two blackbody spectra in format s1, s2, p1, p2
s1 = 0.6
s2 = 0.2
p1 = 0.9
p2 = 1.4
x0 = [s1, s2, p1, p2]

res_pole = minimize (two_BB_diff, x0, args=(Teff, Rns, Mns, sp_pole))
```

```python
print ('Optimised values: (s1, s2, p1, p2) = ', res_pole.x)
print ('Measure of fit quality: ', res_pole.fun)
## Values cited by Yakovlev are: s1 = 0.4613, s2 = 0.2819, p1 = 1.0225, p2 = 1.1914
## Some values coincide and some values differ within 0.1 
## It is impossible to get a better precision because Yakovlev did not describe the way he used to compute the differece between BB spectra and actual integrated spectra
## s1 and s2 can be replaced together with p1 and p2
```

```python
res_eq = minimize (two_BB_diff, x0, args=(Teff, Rns, Mns, sp_eq))
```

```python
print ('Optimised values: (s1, s2, p1, p2) = ', res_eq.x)
print ('Measure of fit quality: ', res_eq.fun)
## Values cited by Yakovlev are: s1 = 0.4807, s2 = 0.2230, p1 = 1.0095, p2 = 1.1891
## Some values coincide and some values differ within 0.1 
## It is impossible to get a better precision because Yakovlev did not describe the way he used to compute the differece between BB spectra and actual integrated spectra
## s1 and s2 can be replaced together with p1 and p2
```

```python
## Let us examine the quality of the fit

eph, spec_2bb = two_BB (res_pole.x, Teff, Rns, Mns)
```

```python
plt.plot (np.log10(eph), np.log10(sp_pole), 'r-',  label=r'Pole red-shifted')
plt.plot (np.log10(eph), np.log10(spec_2bb), 'b:',  label=r'2 BB')



#plt.ylim([15,23])
plt.xlim([-1.2, 0.7])
plt.xlabel(r'$\log\; E$ (keV)')
plt.ylabel(r'$\log\;  H_E^\infty$ (erg s$^{-1}$ cm$^{-2}$ keV$^{-1}$)')
plt.tight_layout()
plt.legend()
```
